### Creating a list to hold the continent for all known countries.

```python
continent_list = []
```

```python
for i in range(11792):
    if W_I_copy['Country'][i] in my_world.get_countries_list():
        continent = my_world.get_country_continent_name(W_I_copy['Country'][i])
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'Channel Islands':
        continent = 'Europe'
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'West Bank and Gaza':
        continent = 'Asia'
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'Virgin Islands (U.S.)':
        continent = 'North America'
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'St. Martin (French part)':
        continent = 'North America'
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'Sint Maarten (Dutch part)':
        continent = 'North America'
        continent_list.append(continent)
    elif W_I_copy['Country'][i] == 'Macao SAR, China':
        continent = 'Asia'
        continent_list.append(continent)
```

```python
print(len(continent_list))
```

### Adding the continent list to the dataset W_I_copy

```python
W_I_copy['Continent'] = continent_list
```
